```python
from langchain_community.llms import Replicate

llm = Replicate(
    model="ibm-granite/granite-3.3-8b-instruct"
)
```

```python
import pandas as pd

df = pd.read_csv("yugioh_deck.csv")
df
```

```python
from langchain_experimental.agents.agent_toolkits.pandas.base import (
    create_pandas_dataframe_agent
)

agent = create_pandas_dataframe_agent(
    llm=llm,
    df=df,
    verbose=True,
    allow_dangerous_code=True,  # opt-in for Python REPL execution
)
```

```python
agent.invoke({"input:" "The Top 20 most used cards per format (e.g., Meta, Non-Meta, Fun, Tournament) with counts."})
```

```python
agent.invoke({"input:" "Find the top-20 most used cards for each deck format."})
```

```python
agent.invoke({"input:" "A grouped bar chart of the Top 20 cards per format (e.g., Non-Meta vs Meta)."})
```
